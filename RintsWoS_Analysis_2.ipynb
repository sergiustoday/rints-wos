```python
#  
import pandas as pd

pd.set_option('display.max_columns', 500)
 
df_anal = pd.read_pickle('RintsWos1_1000.pkl')
df_anal
```

```python
#   
import scipy.stats  as stats
import matplotlib.pyplot as plt
import numpy as np

publ_max = df_anal[2].max() # 885 --      
publ_min = 0
df_sort = df_anal[(df_anal[2] <= publ_max) & (df_anal[2] >= publ_min) & (df_anal[49]  == 1) & (df_anal['cites'] > 0)]
print(df_sort.shape)

x = []
y = []
p = []
for i in range(31):
    rho, pval = stats.spearmanr(df_sort['cites'], df_sort[i])
    x.append(i)
    y.append(rho)
    p.append(pval)
    
df_f_corr3 = pd.DataFrame(list(zip(y, p)), 
               columns =['rho', 'pval']) 

print(df_f_corr3)
    
fig, ax = plt.subplots()

barlist = ax.bar(x, y)
for i in range(31):
    if p[i] > 0.05:
        barlist[i].set_color('r')
        
ax.set_facecolor('floralwhite') #seashell
fig.set_figwidth(18)    #   Figure
fig.set_figheight(9)    #   Figure
fig.set_facecolor('floralwhite')

plt.xticks(np.arange(0, 31, 1)) 
plt.show()
```
