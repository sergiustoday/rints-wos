```python
# Загружаем данные
import pandas as pd

pd.set_option('display.max_columns', 500)
 
df_anal = pd.read_pickle('RintsWos1_1000.pkl')
df_anal
```

```python
# ВСЕ В ЦЕЛОМ
import scipy.stats  as stats
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import numpy as np

publ_max = df_anal[2].max() # 885 -- максимальное число статей в ядре РИНЦ
publ_min = 0
df_sort = df_anal[(df_anal[2] <= publ_max) & (df_anal[2] >= publ_min) & (df_anal[49]  == 1) & (df_anal['cites'] > 0)]
print(df_sort.shape)

x = []
y = []
p = []
for i in range(31):
    rho, pval = stats.spearmanr(df_sort['cites'], df_sort[i])
    x.append(i)
    y.append(rho)
    p.append(pval)
    
df_f_corr3 = pd.DataFrame(list(zip(y, p)), 
               columns =['rho', 'pval']) 

print(df_f_corr3)
    
fig, ax = plt.subplots()

barlist = ax.bar(x, y)
for i in range(31):
    if p[i] > 0.05:
        barlist[i].set_color('white')
        barlist[i].set_edgecolor('black')
        barlist[i].set_hatch('//')
    else:
        barlist[i].set_color('black')
        
ax.set_facecolor('white') #seashell

font = {'family': 'Times New Roman',
        'size': 20,
        }

ax.set_xlabel('Номер показателя в РИНЦ', fontdict = font)
ax.set_ylabel('Корреляция показателя c цитированиями по WoS', fontdict = font)

fig.set_figwidth(18)    #  ширина Figure
fig.set_figheight(9)    #  высота Figure
fig.set_facecolor('white')

plt.xticks(np.arange(0, 31, 1)) 

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(14) 

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14) 
    
legend_elements = [Patch(facecolor='white', edgecolor='black', hatch = '//', label='Корреляция статистически не значима'),
                   Patch(facecolor='black', label='Корреляция статистически значима')]

ax.legend(handles = legend_elements, prop = font)

plt.show()
```

```python
# АНАЛИЗ ВСЕГО В ЦЕЛОМ

# 1) Наибольшее значение корреляции (0,914) у параметра 24 "Число цитирований из зарубежных журналов". Это и не удивительно, так как
#    цитирования из зарубежных журналов включают в себя цитирования из WoS, которые мы и предсказываем. Другое дело, что кроме 
#    журналов из WoS, там ссылки и из других жруналов тоже. И вопрос был в том, насколько эти другие цитирования уменьшат эту
#    корреляцию.
# 2) На втором месте корреляция (0,907) у параметра 8 "Индекс Хирша по ядру РИНЦ". Это можно проинтерпретировать следующим 
#    образом. Индекс Хирша -- хороший показатель, если база, по которому он считается -- хороша. База "Ядро РИНЦ" неплохая. 
#    Для сравнения индексы Хирша по РИНЦ и по elibrary.ru коррелируют значительно хуже. Но, с другой стороны, индексы Хирша даже 
#    по таким слабым базам оказываются лучше многих других показателей. Это говорит о том, что индекс Хирша даже на относительно 
#    слабой базе может получить приемлемый результат.
# 3) На третьем месте (0,888) показатель 14 "Индекс Хирша с учетом только статей в журналах". Это говорит о том, что убирание из 
#    рассмотрения монографий и оставление только лишь статей позволяет сконструировать хороший показатель даже на относительно слабой 
#    базе.
# 5) На четвертом месте (0,870) показатель 5 "Число цитирований из публикаций, входящих в ядро РИНЦ". То, что цитирования по ядру
#    РИНЦ хорошо коррелируют с цитированиями из WoS даже без обработки Хиршом, говорит о том, что ядро РИНЦ -- неплохая база.
# 6) Параметры "число статей" коррелируют всё лучше по elibrary.ru (0,538), по РИНЦ (0,548), по ядру РИНЦ (0,690). 
#    Это аргумент в пользу того, что силы баз расположены именно в таком порядке. Аналогично, параметры "ссылки из статей" по этим 
#    же базам ведут себя также -- 0.767; 0.782; 0.871. Также аналогично индексам Хирша по elibrary.ru (0.786), 
#    по РИНЦ (0.811), по ядру РИНЦ (0.907). То, что разным показателям -- число статей, число ссылок и индексы Хирша -- значения
#    корреляций упорядочено одинаково, это довольно сильные аргументы в пользу того, что РИНЦ действительно сильнее elibrary.ru, а 
#    ядро РИНЦ сильнее РИНЦа. Это, в частности, говорит о том, что процедура исключения недобросовестых журналов из базы РИНЦ
#    работает довольно не плохо.
# 7) Показатели исключительно "интенсивные", завязанные на импакт-фактор 29 и 30 ("Средневзвешенный импакт-фактор журналов, в 
#    которых были опубликованы статьи" и "Средневзвешенный импакт-фактор журналов, в которых были процитированы статьи" 
#.   соответственно) находятся на 9-м и 12-м местах соответственно. Это говорит о том, что в целом, когда в группе есть и очень
#.   плодовитые корифеи исключительно "интенсивные" показатели не являются очень уж сильными предикторами.
#. 8) Единственная отрицательная корреляция у параметра 15 (-0.250). Это выглядит вполне естественно: чем позже исследователь начал
#.   писать научные статьи, тем меньше у него цитирований из WoS.
```

```python
# АНАЛИЗ СЕРЕДНИЧКОВ

# 1) На первом месте также параметр 24, что предсказуемого. И, скорее всего, по тем же причинам.
# 2) А вот на втором месте параметр 29 "Средневзвешенный импакт-фактор журналов, в которых были опубликованы статьи". Это уже
#.   интересно. Это исключительно интенсивный показатель, не содержащий никакой экстенсивной составляющей. Он переиграл (хотя и не 
#.   сильно) все возможные Хирши. Это можно попытаться объяснить тем, что базы, по которым считаются Хирши, весьма разнородные.
#.   А тут импакт-фактор в чистом виде. Это более сильный предиктор, если экстенсивная составляющая не так сильно велика.
# 3) На третьем месте параметр 30 "Средневзвешенный импакт-фактор журналов, в которых были процитированы статьи". Это тоже 
#.   импакт-фактор, и факт того, что он тоже переигрывает разного рода экстенсивные параметры, говорит о том, что для крепких 
#.   середнячков учет плотности дискурса может быть важнее некоторых экстенсивных параметров.
# 4) На четвертом месте параметр 5 "Число цитирований из публикаций, входящих в ядро РИНЦ". Здесь стоит отметить два момента.
#.   Во-первых, показатель основывается на ядре РИНЦ, что в очередной раз говорит, ядро РИНЦ -- неплохая база. А во-вторых, на этой
#.   выборке параметр 5 опережает параметр 8 "Индекс Хирша по ядру РИНЦ". Это позволяет предположить, что для не архиплодовитых
#.   авторов, возможно, стоит ориентироваться на число ссылок из сильной базы. А вот для корифеев уже (и это будет некоторым 
#.   образом подтверждено ниже) имеет смысл ориентироваться на индекс Хирша по сильной базе.
# 5) Мы видим, что есть четыре параметра -- 20, 21, 25 и 26 -- по которым наблюдается небольшая, но тем не менее существенная 
#.   статистически значимая отрицательная корреляция. Наибольшей по абсолютному значению отрицательной корреляцией обладает параметр
#.   20 "Число статей в российских журналах" (-0.526). Получается, что избыточная гонка за числом статей в не международных журналах
#.   идет вразрез с получением ссылок из WoS. 
# 6) На втором месте с конца параметр 25 "Число цитирований из российских журналов" (-0.472)
#.   Видим, что при прочих равных цитирования -- часто менее манипулируемый параметр, чем число статей.
# 7) На третьем месте с конца параметр 21 "Число цитирований из российских журналов из перечня ВАК" (-0.432). Это говорит о том, 
#    что хоть и некоторые усилия по формированию перечня ВАК видны (всё-таки отрицательная корреляция меньше по абсолютному значению,
#.   чем у просто российских журналов), но тем не менее эти усилия оказываются недостаточными, чтобы база ВАК могла считаться 
#.   хорошей: как и в случае с российскими журналами, ориентация на ВАКовский список скорее уводит в сторону от международного 
#.   признания, чем помогает продвинуться вперед.
# 8) На четвертом месте с конца параметр 26 "Число цитирований из российских журналов из перечня ВАК" (-0.395). Здесь можно сделать
#.   те же выводы, что и раньше. Во-первых, цитирования при прочих равных оказывается менее манипулируемым параметром, чем число 
#.   статей. И во-вторых, ВАКовский список несколько чище, чем перечень всех российских журналов, но всё же не настолько, чтобы на него
#.   можно было ориентироваться. Более того, ориентироваться на него просто вредно.
# 9) Видим, что параметр 15 "Год первой публикации" статистически значимо не коррелирует с числом ссылок по WoS. Это говорит о том, 
#    что для этой группы еще не очень возрастных середнячков не так важно, чуть раньше или чуть позже автор начал писать. Важно, 
#.   в журналах с каким импакт-фактором он публикуется.
```
